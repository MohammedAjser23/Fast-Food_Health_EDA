### 1.4.1 Creation of Health Category (BMI-Based Classification)

Although BMI values are numerically precise, interpreting them directly in visualizations can be less intuitive. To improve readability and categorical analysis, BMI values were grouped into standard health classifications.

The categories created include:

+ Underweight  
- Normal  
+ Overweight  
- Obese  


> *Instead of analyzing continuous BMI values alone, categorized BMI provides structured health segmentation for deeper pattern analysis.*

```python
# Defining a simple function to categorize BMI
def get_category(bmi):
    if bmi < 18.5:
        return 'Underweight'
    elif 18.5 <= bmi < 25:
        return 'Normal'
    elif 25 <= bmi < 30:
        return 'Overweight'
    else:
        return 'Obese'

# Applying the function to create a new column
df['Health_Status'] = df['BMI'].apply(get_category)

# Showing the new column
df[['BMI', 'Health_Status']].head()
```

This transformation serves the following purposes:

* *Simplifies comparison across individuals*  
* *Enables grouped visualizations*
* *Aligns with global health standards*  
* *Improves interpretability for non-technical audiences*

### 1.4.2 Categorization of Weekly Fast-Food Consumption

Weekly fast-food intake was originally recorded as numerical frequency counts. Weekly fast food intake was converted from numerical
counts into consumption levels to simplify comparison across individuals.

To improve analytical clarity, fast-food frequency was grouped into consumption levels:

+ Low Consumption  
- Moderate Consumption  
+ High Consumption  

>*By converting continuous frequency into structured behavioral levels, the dataset becomes more suitable for comparative and categorical analysis.*

```python
# Create fast food consumption level
def fast_food_level(meals):
    if meals <= 3:
        return "Low"
    elif meals <= 7:
        return "Moderate"
    else:
        return "High"

df['Fast_Food_Level'] = df['Fast_Food_Meals_Per_Week'].apply(fast_food_level)

df[['Fast_Food_Meals_Per_Week', 'Fast_Food_Level']].head()
```

This categorization was performed to:

* *Simplify behavioral comparisons*  
* *Improve grouped visualizations*  
* *Identify consumption patterns more clearly*  
* *Strengthen correlation analysis with BMI and health score*

  >📌 It is important to note that not all outliers represent data errors. Some extreme values may reflect genuine health conditions. Therefore, outlier handling decisions must be made carefully to avoid removing meaningful observations.

## 2.1 Gender Distribution Analysis

The first step in exploratory analysis is understanding the demographic composition of the dataset. Gender distribution provides foundational context for interpreting subsequent health and consumption patterns.

This visualization helps:

* Identify dataset balance  
* Prevent demographic bias in interpretation  
* Provide context for gender-based comparisons later  

*A count-based visualization is used to observe how participants are distributed across gender categories.*
>+ *If one gender is over-represented, it could lead to 'Gender Bias' in my final conclusions.*

```python
plt.figure(figsize=(8,4))   # width, height in inches
df['Gender'].value_counts().plot(kind='bar')
plt.title('Gender Distribution')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

>    **Interpretation**:                         
    The bar chart shows that participants are distributed across gender categories in a fairly balanced manner. This is important because it means my analysis of fast food and BMI won't be heavily biased toward just one group, making it suitable for gender-based comparisons.

## 2.2 BMI Distribution Analysis

Body Mass Index (BMI) distribution is analyzed to understand the overall health composition of the participants.

This analysis helps:

* Identify concentration of individuals in specific BMI ranges  
* Observe skewness or clustering patterns  
* Establish baseline health understanding  

*A *histogram* with density curve is used to visualize the distribution.*

```python
plt.figure(figsize=(8,4))

# 'histplot' creates the bars, 'kde=True' adds the smooth wavy line over them
sns.histplot(df['BMI'], kde=True, color='skyblue')

# Adding labels
plt.title('How BMI is Distributed Among Participants')
plt.xlabel('BMI Value')
plt.ylabel('Number of People')
plt.show()
```

>    **Interpretation**:  
    The BMI distribution shows that most individuals fall within a mid-range BMI category, with fewer individuals at the very low or very high extremes. This indicates a relatively balanced spread, even there are several people with a BMI over 30, which is considered obese.

## 2.3 Fast Food Consumption Frequency per Week

Understanding how often participants consume fast food is essential before analyzing its health impact.

This visualization helps:

* Identify dominant consumption patterns  
* Distinguish between low and high intake groups  
* Establish behavioral baseline  

A count-based frequency visualization is used.

> **Now I want to see the 'cause'. How many fast food meals are these people eating every week?**



```python
plt.figure(figsize=(8,4))

# 'countplot' simply counts how many times each number appears in the column
sns.countplot(x='Fast_Food_Meals_Per_Week', data=df, palette='viridis')
plt.title('Number of Fast Food Meals Eaten per Week')
plt.xlabel('Meals Per Week')
plt.ylabel('Count')
plt.show()
```

>    **Interpretation**:  
    The chart shows that eating fast food 2 to 4 times a week is the most common habit. Very few people eat it 0 times, while fewer individuals fall at the extreme low or high ends. This gives us a good 'baseline' of consumption habits within the dataset.

## 2.4 Relationship Between Fast Food Consumption and BMI

This visualization investigates whether an observable relationship exists between fast-food intake and BMI levels.

This analysis helps:

* Detect positive or negative trends  
* Evaluate potential health impact  
* Support or challenge the research hypothesis  


>This is the most important part of the project, **Does eating more fast food
actually lead to a higher BMI?**


```python
plt.figure(figsize=(8,4))
plt.scatter(df['Fast_Food_Meals_Per_Week'], df['BMI'])
plt.title('Does More Fast Food Mean a Higher BMI?')
plt.xlabel('Fast Food Meals Per Week')
plt.ylabel('BMI')
plt.show()
```

As the number of fast food meals per week increases (moving to the right), the BMI scores also tend to move higher (moving up). This suggests a strong connection between fast food and weight gain.

This visualization analyzes whether fast-food consumption behavior differs across gender categories.

This step is important to:

* Detect behavioral differences  
* Identify demographic consumption patterns  
* Support comparative analysis  

A grouped visualization is used to compare frequency levels.

## 2.7 Physical Activity vs BMI

Physical activity is a key moderating variable in health analysis. This visualization explores whether higher activity levels correspond with lower BMI values.

This helps:

* Evaluate lifestyle impact  
* Identify protective behavioral factors  
* Strengthen health interpretation  

A relational plot is used to examine interaction patterns.

>**We are looking for a downward trend—does more exercise lead to a lower BMI?**


```python
# Group data by physical activity hours and calculate mean BMI
activity_bmi = df.groupby('Physical_Activity_Hours_Per_Week')['BMI'].mean()
plt.figure(figsize=(8, 4))
plt.plot(activity_bmi.index, activity_bmi.values, marker='o')
plt.title('Average BMI vs Physical Activity Hours per Week')
plt.xlabel('Physical Activity Hours per Week')
plt.ylabel('Average BMI')
plt.show()
```

>    **Interpretation:**    
    The line chart indicates a general downward trend in average BMI as physical activity hours increase. This confirms that participants who spend more time exercising tend to maintain a lower, healthier BMI, highlighting the importance of regular exercise in maintaining a healthy body weight.



## 2.8 Trend Analysis – Fast Food Consumption vs Health Score

This section investigates whether increasing fast-food consumption corresponds to changes in overall health score.

This analysis helps:

* Evaluate direct behavioral impact  
* Identify linear or non-linear trends  
* Support causal interpretation discussion  

*A line trend visualization is used to observe directional patterns.*

```python
# Prepare grouped data
fastfood_bmi_trend =(df.groupby('Fast_Food_Meals_Per_Week', as_index=False)['BMI'].mean())
plt.figure(figsize=(8,4))
sns.lineplot(data=fastfood_bmi_trend, x='Fast_Food_Meals_Per_Week', y='BMI', marker='o')
plt.title('Trend of Average BMI vs Fast Food Meals Per Week')
plt.xlabel('Fast Food Meals Per Week')
plt.ylabel('Average BMI')
plt.show()
```

>    **Interpretation:**                        
    There is a sharp downward trend. Participants who eat fast food 5+ times a week report significantly lower health ratings (1 or 2) compared to those who eat it rarely.This proves that diet has a direct impact on how healthy a person feels.

## 2.9 Overall Health Score Distribution

Understanding the distribution of health scores provides insight into general population well-being.

This visualization helps:

* Identify concentration ranges  
* Detect skewness  
* Support interpretation of trend results  

*A distribution plot is used to observe score spread.*

```python
# Create health score categories
health_score_counts = df['Overall_Health_Score'].value_counts().sort_index()
plt.figure(figsize=(7,5))
plt.pie(health_score_counts,
        labels=health_score_counts.index,
        autopct='%1.1f%%',
        startangle=90)
plt.title('Distribution of Overall Health Scores')
plt.axis('equal')
plt.show()
```

>    **Interpretation:**                            
    This chart shows the "health split." The largest slices are for,the rating with group is moderately healthy. It helps us identify if the majority of the study group falls into the "at-risk" category. This provides a quick overview of general health levels before deeper multivariate analysis.

## 2.10 Correlation Analysis Between Numerical Variables

To quantify relationships between key numerical variables, correlation analysis is performed.

This analysis helps:

* Measure strength and direction of relationships  
* Validate visual trends statistically  
* Identify strongest predictive associations  

>**This step examines the correlation between key numerical variables such as
fast food consumption, BMI, physical activity, sleep hours, and overall health score.**

```python
# Select only numerical columns
numerical_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
correlation_matrix = numerical_df.corr()
plt.figure(figsize=(10, 7))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Health & Lifestyle Variables')
plt.show()
```

>   🔎  **What this shows:**
    *This heatmap is the ultimate summary. High positive numbers (red) show variables that increase together, like Fast Food and BMI. High negative numbers (blue) show things that decrease each other, like Fast Food and Health Rating. This mathematically proves our earlier visual findings.*
